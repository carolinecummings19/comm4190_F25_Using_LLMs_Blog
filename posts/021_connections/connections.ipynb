# December 18 NYT Connections Puzzle (WARNING: Spoilers)

After completing the New York Times Connections this morning in a record time, I was curious to see if an LLM could do the same. I first prompted ChatGPT 5.2 with a simple prompt of 

<div style="background-color:#f6f8fa; border-left:4px solid #0366d6; padding:12px; font-family: 'Fira Code', monospace; font-size: 0.90rem;">
Can you solve this connections puzzle?
</div>

{width=40%}


The output was:
> Here you go  the four groups:
> 
> * **Preceded by SOCIAL**: **BUTTERFLY, MEDIA, SECURITY, WORKER**
> * **Main character / leading figure**: **HERO, LEAD, PRINCIPAL, STAR**
> * **Imagination / illusion (things in your head)**: **DAYDREAM, FANCY, FANTASY, ILLUSION**
> * **Craft supplies**: **FELT, GLITTER, PIPE CLEANER, POMPOM**


On the first try, ChatGPT gave me the correct categorizations. In order to make it a bit harder, I tried to then ask it to guess the color of the categorization. These colors are based on level of difficulty starting with the easiest of yellow, then green, then blue, then the hardest of purple. I prompted with the following:

<div style="background-color:#f6f8fa; border-left:4px solid #0366d6; padding:12px; font-family: 'Fira Code', monospace; font-size: 0.90rem;">
Can you solve this connections puzzle? Label what you think is the yellow, green, blue, and purple categories.
</div>

ChatGPT's response:

> **Yellow (Craft supplies):** FELT, GLITTER, POMPOM, PIPE CLEANER
> 
> **Green (Social ___ phrases):** BUTTERFLY, MEDIA, SECURITY, WORKER
*(social butterfly / social media / social security / social worker)*
> 
> **Blue (Main character / top-billed person):** HERO, LEAD, PRINCIPAL, STAR
> 
> **Purple (Things you can imagine / not real):** DAYDREAM, FANCY, FANTASY, ILLUSION


For good measure, I also compared the output with Gemini with the exact same prompt. Here is what Gemini's final response was:

> **Final Confirmed Set:**
> 
> * **Yellow:** FELT, GLITTER, PIPE CLEANER, POMPOM (Crafts)
> * **Green:** HERO, LEAD, PRINCIPAL, STAR (Lead Roles)
> * **Blue:** DAYDREAM, FANCY, FANTASY, ILLUSION (Types of mental imagery/Whimsy)
> * **Purple:** BUTTERFLY, MEDIA, SECURITY, WORKER (Words following "SOCIAL")

{width=50%}

Again, each of the prompts led to the correct groupings of the words. However, both LLMs incorrectly identified the difficulty level that the creators of the NYT Connections placed on each of the categories. Both LLMs thought the crafts categories was the easiest when the actual game placed it as third. ChatGPT thought that Social __ was the second hardest while Gemini correctly placed it as the hardest. I would say from experience that usually when its a category of word __ is the hardest. Perhaps Gemini has some training data that specifies this continuity and is why it is correct. Gemini also correctly identifies the second level. 

This highlights an interesting distinction: LLMs can solve the puzzle efficiently, but they do not fully replicate the cognitive friction humans experience while solving it. While they outperform me in speed and accuracy, they lack the lived intuition that makes certain categories feel deceptively difficult. In that sense, NYT Connections reveals not just how capable these models are, but also where human reasoning still diverges from machine pattern matching.
