# Nano Banana for Image Editing

### Gemini 2.5 Flash Image model released in August of this year

Nano Banana is a new creative AI tool known for subtle, high-quality edits that keep images looking natural rather than over-processed. I was honestly skeptical when I first went to use it. From past experience, I’ve noticed that many image models struggle with consistency — they tend to change facial features or add an unnatural “AI-generated” look. The last thing I wanted was a headshot that didn’t really look like me.

My prompt was:

“Can you make the shadows in this image less dramatic? I would like my face to be a bit lighter and the overall colors in the image to be slightly warmer tones. Keep the Penn Engineering banner and how I look the same.”



Nano Banana softened the harsh shadows on my face, brightened my features just enough, and warmed up the tones — all while keeping me looking like myself and preserving the Penn Engineering banner in the background.

The difference might seem subtle at first glance, but it completely elevates the professionalism of the photo. I was impressed not just by the technical accuracy, but by how natural the edit feels. It doesn’t scream “AI edit” at all — it just looks like a much better headshot.

I’ll be updating my LinkedIn profile with the improved version, and I can confidently say Nano Banana exceeded my expectations.
