# Using Gemini 2.5 Flash to Flag Potential Spam

I grabbed a few sketchy-looking emails (and one that I thought was real) and asked Gemini 2.5 Flash to score them for risk and explain why. Below are the screenshots of the emails and model replies to the following prompt.

**Prompt**

```
Is this likely marketing or phishing? Rate the spam/phishing risk (0100) and list concrete indicators (domain, spelling, links, urgency). Keep response to under 150 words.
```

## Example 1  Free Roof Upgrade

{width=50%}

**Response**

<aside style="background:#f6ffed;border:1px solid #b7eb8f;padding:10px;border-radius:10px">

**Assessment: Marketing (High Spam Risk)**
This email is a **lead generation spam** attempt, rated a **65/100** risk. It uses deceptive tactics to harvest personal information for sales calls, though it is not a direct phishing attack for passwords.

**Key Indicators:**

* **Sender:** Gibberish email address (`gyjeohdhgl@...`) shows it's not from a legitimate, verifiable company.
* **Urgency:** The subject line ("Claim It Now!") and language create a false sense of a time-sensitive, personalized offer.
* **Deceptive Links:** Links like "Check Eligibility" and "Secure link | 100% privacy protected" are manipulative, aiming to get you to click and submit personal data.

**Action:** This is a high-risk spam that should be deleted.
</aside>



## Example 2  Prime Payment Failed

{width=50%}

**Response**
<aside style="background:#f6ffed;border:1px solid #b7eb8f;padding:10px;border-radius:10px">
    
**Assessment: PHISHING (Critical Risk)**

This is a **phishing** attempt, rated a high **90/100** risk, designed to steal your credit card details and login credentials by impersonating Sirius XM.

**Critical Indicators:**

* **Fake Sender:** The email address is random gibberish (`fvgvcankatx@...`), not an official Sirius XM address.
* **Extreme Urgency:** The subject line, "**Your Account... Will Be Removed Today**," pressures immediate action to prevent loss of service.
* **Deceptive Offer:** It offers "90 days for FREE" to bait a click, followed by the requirement to enter credit card details for "validation."
* **Malicious Link:** The "RENEW SUBSCRIPTION NOW" button directs you to a fraudulent website to harvest your personal and financial information.

**Action:** This is highly malicious. Do **not** click the link and delete the email immediately.
</aside>
