# Planning my time in Zaragoza, Spain

My parents, my grandma, and I are going to Spain in January and have not planned the portion of the trip where we are in Zaragoza. We are visiting my mom's cousin there and want to hit the major landmarks. To help my mom plan, I decided to enlist Gemini to give us recommendations to look into. Here is the prompt I gave it 

<div style="background-color:#f6f8fa; border-left:4px solid #0366d6; padding:12px; font-family: 'Fira Code', monospace; font-size: 0.90rem;">
I am planning a trip to Spain and have 4 days in Zaragoza. We already have accommodations in the city. We get there on January 18th in the afternoon and leave on the 22nd in the morning. I am not too familiar with the city so would like recommendations on how to fill our time there. Keep in mind we do not have a car but could potentially rent one. Here was Gemini's Response:

> **Day 1: Arrival & The "El Tubo" Experience**
> * **Afternoon Arrival:** Settle into your accommodation. If you are near the center, start with a walk to Plaza del Pilar, one of the largest pedestrian squares in Europe.
> * **Evening:** Head to **El Tubo**, a famous maze of narrow alleys filled with tapas bars.
>    * *Tip:* Try the famous mushrooms at *El Champi* or the croquettes at *Doña Casta*.
> 
> **Day 2: Mudéjar Architecture & Roman Roots**
> * **Morning:** Visit the **Basilica del Pilar**. It’s free to enter, but pay the small fee to take the elevator up the tower for the best panoramic view of the city and the Ebro River.
> * **Late Morning:** Walk 2 minutes to **La Seo Cathedral**. While the Basilica is more famous, many locals prefer La Seo for its stunning mix of Gothic and Mudéjar (Islamic-influenced) styles.
> * **Afternoon:** Follow the **Roman Route**. Zaragoza was the Roman city Caesaraugusta. Visit the **Roman Theatre** (the most impressive ruin) and the **Forum Museum** located underground.
> 
> **Day 3: The Aljafería & Modern Art**
> * **Morning: Aljafería Palace**. This 11th-century Islamic palace is a UNESCO World Heritage site and arguably the most important Moorish building outside of Andalusia.
>   * *Note:* It currently houses the regional parliament, so booking tickets online in advance is highly recommended.
> * **Afternoon:** Visit the *Goya Museum*. Francisco de Goya was born just outside the city, and this museum houses a significant collection of his engravings and paintings.
> * **Alternative:** If you prefer modern art, the *IAACC Pablo Serrano* is a striking building with great contemporary exhibits.
> 
> **Day 4: Day Trip (No Car Needed)**
You can easily see incredible sites outside the city using Spain's efficient train (AVE) or bus network:
> * **Option A:** **Alquézar (Bus/Tour)**: Often cited as one of Spain's most beautiful medieval villages, famous for its "footbridges" walk along the river.
> * **Option B:** **Monasterio de Piedra (Bus)**: A stunning natural park filled with waterfalls and a 12th-century monastery. There is a direct bus from the Zaragoza Delicias station that takes about 2 hours.
> * **Option C:** **Belchite (Bus)**: For history buffs, the ruins of "Old Belchite" (destroyed in the Civil War) serve as a haunting "ghost town" memorial.
