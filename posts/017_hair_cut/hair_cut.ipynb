# How I Used ChatGPT-5 to Cut My Roommate’s Hair

### I actually (successfully) cut someone’s hair with an LLM as my live coach.

Cutting hair is one of those things that seems simple until you’re the one actually holding the scissors. My roommate asked me to trim and layer her long, wavy hair and I told her I'd give it a go.

So naturally, I opened ChatGPT-5 and asked it how to guide me step-by-step.

## User Prompt 1

<div style="background-color:#f6f8fa; border-left:4px solid #0366d6; padding:12px; font-family: 'Fira Code', monospace; font-size: 0.95rem;">
My friend asked me to cut her hair. How would you recommend I do it? She has long hair that’s slightly wavy with layers
</div>

### ChatGPT 5 Response 1

ChatGPT gave a beginner-friendly overview: tools needed, the “golden rules,” and a safe method using sectioning or the “unicorn cut.” It emphasized trimming very conservatively, keeping symmetry, and using point-cutting to soften the ends.

It told me not to use kitchen scissors. Unfortunately, those were the only scissors I had.

---

I uploaded a picture of my roommate’s hair from the back.
I clarified that I didn’t want to do the unicorn cut.

## User Prompt 2

{width=60%}

<div style="background-color:#f6f8fa; border-left:4px solid #0366d6; padding:12px; font-family: 'Fira Code', monospace; font-size: 0.95rem;">
I don’t think I want to do a unicorn cut so I think I’ll section it and cut it.
</div>

## ChatGPT-5 Response

ChatGPT analyzed the photo and confirmed:

The hair type (long, wavy)

Uneven ends

Recommended a 3-section method

Told me exactly how to trim the bottom, middle, and top sections safely

It was like having a virtual hairstylist supervise.


---

I sent another picture after I clipped her hair into multiple sections.

## User Prompt
{width=60%}

### ChatGPT-5 Response

It outlined the steps to go from the current photo including what to do for the bottom, middle, and top sections and then doing the side + face-framing and cross-check. It also made the comment:

> ## Your Sectioning in the Photo is Perfect for Controlled Layers
> 
> Just keep these rules in mind:
> 
> * Bottom = blunt(ish) trim
> * Middle = longest/softest layers
> * Top = lightest, shortest layers
> * Point-cut everything except the bottom baseline

It adjusted its instructions based on the exact setup in the image.

--- 

I did the length trim and sent another picture of the result.

## User Prompt
{width=60%}

<div style="background-color:#f6f8fa; border-left:4px solid #0366d6; padding:12px; font-family: 'Fira Code', monospace; font-size: 0.95rem;">
Now I want to do layers.
</div>

## ChatGPT-5 Response

This is where ChatGPT-5 really shined.
It told me exactly how to layer based on the photo:

Release the middle section

Cut the “guide” layer

Elevate at 45°

Trim only 1/2 inch

Use vertical point-cutting to avoid harsh lines

Then move to the top section and add only light layering

---

The layering was the part that I was most worried for but I followed the instructions and think my roommates hair turned out great!

Would I trust ChatGPT to cut my own hair?
... Probably not.

Would I use it again to cut someone else’s hair?
Absolutely.
