I thought this book was an interesting read and my first book on AI. I appreaciated the historical bits where the authors looked back on certain historical events. There were a lot of claims surrounding how historical events panned out and how AI will do the same. However, these claims are very broad and some were not very similar to the current state and innovation of AI so didn't seem very meaningful. 

I also thought the authors were super optimist about the outcomes of the AI boom, sometimes too much so. Reid Hoffman has a unique perspective coming from being a billionaire and on the board of Microsoft which probably makes his positivity understandable. However, I think for most of the readers his broad claims of how AI will improve society and permissionless innovation will lead to safer models is not something that may resonate with the average person. Some of it seemed pretty dystopian too.

I also thought this book was overly U.S. centric and also the title could be a little misleading considering it did not really talk about AI agents but rather human agency. It sort of seemed like a manifesto on adopting AI in the hopes the rapid innovation will correct any issues with current models.

Overall, thought it was fairly easy to read and interesting food for thought. It definitely wasn't the greatest book I've read and I certainly did not agree with a lot in the book but not that bad.
