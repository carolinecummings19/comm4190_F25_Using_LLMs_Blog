# Review of *Superagency* Chapters 3-4

## Chapter 3: What Could Possibly Go Right?

This chapter focuses on the potential of AI using specific examples and how its benefits will depend on how it is implemented. There are specific examples mentioned in the education space, healthcare, and being more productive. 

Specifically, they talk about mental health chatbot, WoeBot. They discuss how the company want to increase access to mental health support using LLMs. The authors discussed some pitfalls with this technology but that when implemented effectively, AI can provide a lot of value to society. 

They encourage readers to change their fear into looking into the opportunity AI can provide. By mentioning all these different applications of AI in different realms of life, the authors are trying to build up their arguement of how AI agents can boost individual productivity, well-being, and agency, for a more abundant and capable society.

## Chapter 4: The Triumph of the Private Commons

Over the course of this chapter, the authors define a "private commons" as valuable services that are 'free' from private companies (i.e. Google maps, Dropbox, etc.) in exchange for user data. They frame the classic LLMs offered by OpenAI and Google as a type of private common. 

The authors then compare the nature of these private commons to a traditional common, like a natural resource. Specifically, they talk about 'tragedy of the commons' dilemma which is a concept often used to describe limited resources humans tap in to. It states that shared resources are inevitably ruined by overuse because each party wants to maximize their own profit or benefit. In the context of digital information, the authors claim that there is an opposite effect: users who utilize the private commons more benefit themselves and help to improve the private commons by giving them their data. More users leads to better value provided by the resource. 

In the context of AI, the authors argue the same idea applies to AI. These AI private commons can amplify human capabilites, educational opportunities, and more fluid social mobility the more people use them. This framework stages AI as something positive for societal progress.

## Overall thoughts

Both of these chapters are building up the idea of AI's potential to positively impact society. Thus, a lot of history and context about how society used to function and the current state of society is mentioned. I quite enjoy these bits because I am learning a lot about where gaps in society are (mental health resources, wealth gaps, etc). By connecting AI to long-standing gaps in societysuch as access to mental health care, education, and economic mobility, it makes the discussion feel more grounded and less speculative. It helped me better understand where AI could realistically make a difference if implemented responsibly.

The tone of these chapters feels very aligned with a bloomer perspective. They emphasize that AI is not guaranteed to improve society, but it has the potential to do so if humans remain thoughtful of the technology. This may resonate strongly with readers who are already optimistic about technological progress, but could be challenged by more skeptical or gloomer audiences who are concerned about power imbalances, data ownership, or corporate control.

I found these chapters compelling and motivating. I still felt like there were still open questions about who actually utilizes the private commons and whether or not it is truly closing the societal gap or widening it. Some of the use cases mentioned have a strong assumption that everybody has equal access to these technologies which is simply not true.
