# Review of *Superagency* Chapters 7-8

## Chapter 7: Informational GPS

This chapter discusses how LLMs can act as a GPS to navigate the massive amount of information available around us. This includes informing any decision making, verifying information correctness, and increasing creativity and productivity. This chapter emphasizes again how AI can give us better opportunities when leveraged effectively and should not be feared. 
 

## Chapter 8: Law is Code

Chapter 8 discusses a unique intersection between technology and governance systems. Specifically, it talks about how techology is becoming increasingly intertwined with these legal systems including law and regulations. Additionally, it mentions how AI may lead us to rethink and refactor our existing human systems because of outdated beliefs and ethics. The chapter once agains gives an example of how AI and humans can work together to improve society as a whole.
