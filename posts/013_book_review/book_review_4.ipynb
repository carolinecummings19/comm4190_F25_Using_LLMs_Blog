# Review of *Superagency* Chapters 7-8

## Chapter 7: Informational GPS

This chapter discusses how LLMs can act as a GPS to navigate the massive amount of information available around us. This includes informing any decision making, verifying information correctness, and increasing creativity and productivity. This chapter emphasizes again how AI can give us better opportunities when leveraged effectively and should not be feared. 

A metaphor I found interesting in the book was "knowledge fog" which described how the current state of the internet is an information overload or a fog. Google give you a list of links to information thats practically endless. The authors claim that LLMs can act as a path through the fog by synthesizing large amounts of data. 

One point I didn't quite agree with was using LLMs as a truth verifier. I think that LLMs are still not at the point where I trust the output enough to claim it is true on topics I'm not well-versed in. LLMs still hallucinate and can provide misinformation in some cases. However, I think they are great resources to converse with so could help individuals make more informed decisions.

## Chapter 8: Law is Code

Chapter 8 discusses a unique intersection between technology and governance systems. Specifically, it talks about how techology is becoming increasingly intertwined with these legal systems including law and regulations. Additionally, it mentions how AI may lead us to rethink and refactor our existing human systems because of outdated beliefs and ethics. The chapter once agains gives an example of how AI and humans can work together to improve society as a whole.

The authors mention how "law is code" which implies that legal agreements could eventually be self-executing. They also mention how AI could help produce a more proactive system through computer vision, turning minor violations like public intoxication or illegal parking into zero-tolerance offenses. This, to me, felt a bit dystopian. I think our laws right now disproportionately affect certain demographics because of societal biases and deep-seated discrimination. By enforcing laws even more, specific groups of people will likely be affected much more than others. I am a proponent of using AI to fix these societal issues before having them enforce any laws.
