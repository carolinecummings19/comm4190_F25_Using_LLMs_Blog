# Review of Superagency Chapters 1-2

## Introduction

The book starts off with a bit of history on technologies and their role in humanity. They talk about the printing press, the telephone, the Model T, etc and how people have always had skepticism and negative feelings towards these emerging technologies, claiming that they will replace humans in industry or make people less productive in society. In contrast, society has lower unemployment rates and more productivity today. They make claims of how technological progress is inevitable because of the innate collaborative and competitive nature of humans. I thought this was a good opening that set the scene of how people have reacted in the past so that the reader can see the connection with how certain people feel about AI today. 

## Chapter 1: Humanity Has Entered the Chat

Now, the authors set the stage of post-COVID-19 chaos, with inflation concerns and big tech lay-offs. Then, in November of 2022 OpenAI released ChatGPT without advanced notice or hype but became popular because of it's unprecedented performance: versatile and incredibly knowledgable. It created a wave of urgency within the tech space because big tech companies felt they fell behind. The government also became skeptical and held hearings on AI. Then, a tangent on how AI works and undermines the true "intelligence" it has.  

The authors also cover how these AI systems, namely ChatGPT, were adopted very quickly and had immediate impact on society. The main ideas covered human control and agency and how AI systems can increase agency when used efficiently. AI should enhance human decisions and not replace them and we are the people who should take that initiative.

## Chapter 2: Big Knowledge

This chapter drew on ideas from the book *1989* by George Orwell and concerns over technological surveillance and how they have always existed. The authors highlighted how these concerns have often been exagerated and also incorporated a real story of fears over a data center managed by the goverment in the 1960s. They also cited that access to computers and the internet led to more individual empowerment rather than opression. 

I enjoyed this chapter because I was learning new things about the history of technology in the U.S. that I didn't know previously. I also think this was a pretty strong argument by citing past examples of people having fears of new technology but ended up being fine. The past fears of government control ended up being overpowered by the democratization of knowledge and information from the people. 

## Overall Thoughts

So far, I think the book does a great job of connecting past reactions to new technologies with current attitudes toward AI. Ive enjoyed how it uses historical context to show that fears about innovation often fade as society adapts and benefits from progress. Overall, its made me think more critically about how human agency and collaboration shape the role of technology in our lives.
