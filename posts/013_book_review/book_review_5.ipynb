# Review of *Superagency* Chapters 9-11

## Chapter 9: Networked Autonomy

Chapter 9 talks about how AI will support more complicated human endeavors by enhancing human agency. This teamwork will lead to even greater impact than before because of AI being a supertool. One specific concept mentioned was an "agentic AI mesh" where multiple AI agents work together without much human involvement. These autonomous systems will require good guardrails and guidelines in order to stay on track and be productive in technological innovation. 

## Chapter 10: The United States of A(I)merica

This chapter highlights the need for the U.S. to prioritize AI so that it can stay a world leader. The government can utilize AI to modernize their governance, improve public services, and drive economic growth. It relates AI to other national infrastructure like highways or the internet which can put the U.S. in another era of prosperity. The end of the chapter emphasizes the importance of being cautious when implementing AI in the government because of risks of job displacement, monopoly, and ethical use. The authors frame AI as a tool that America needs to invest in for the betterment of the nation.

## Chapter 11: You Can Get There from Here

This chapter is a culmination of themes previously mentioned in the book. Specifically, how AI can help humans build a better future and how people should actively participate in shaping AI's trajectory. A large portion of the chapter is also on certain challenges that should be guarded against like ethical issues, amplifying inequalities, and corosion of critical thinking.
