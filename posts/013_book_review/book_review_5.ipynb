# Review of *Superagency* Chapters 9-11

## Chapter 9: Networked Autonomy

Chapter 9 talks about how AI will support more complicated human endeavors by enhancing human agency. This teamwork will lead to even greater impact than before because of AI being a supertool. One specific concept mentioned was an agentic AI mesh where multiple AI agents work together without much human involvement. Hoffman suggests that this "mesh" allows individual humans to act as the "CEO" of their own personal department of AI experts. These autonomous systems will require good guardrails and guidelines in order to stay on track and be productive in technological innovation. Ultimately, the goal is to move from AI as a simple assistant to AI as a collaborative network that can handle complex, multi-step projects independently.

## Chapter 10: The United States of A(I)merica

This chapter highlights the need for the U.S. to prioritize AI so that it can stay a world leader. The government can utilize AI to modernize their governance, improve public services, and drive economic growth. It relates AI to other national infrastructure like highways or the internet which can put the U.S. in another era of prosperity. The authors argue that if the U.S. fails to lead in AI development, it risks ceding the global standard for ethics and safety to other nations with different values. The end of the chapter emphasizes the importance of being cautious when implementing AI in the government because of risks of job displacement, monopoly, and ethical use. The authors frame AI as a tool that America needs to invest in for the betterment of the nation, suggesting that a "National AI Strategy" is as vital to the 21st century as the Space Race was to the 20th.

## Chapter 11: You Can Get There from Here

This chapter is a culmination of themes previously mentioned in the book. Specifically, how AI can help humans build a better future and how people should actively participate in shaping AIs trajectory. Hoffman emphasizes that the future isn't something that just "happens" to us, but something we co-author through our choices today. A large portion of the chapter is also on certain challenges that should be guarded against like ethical issues, amplifying inequalities, and corrosion of critical thinking. The book concludes with an optimistic call to action, reminding readers that while the risks are real, the greatest risk of all is failing to use AI to solve our most urgent global problems.
