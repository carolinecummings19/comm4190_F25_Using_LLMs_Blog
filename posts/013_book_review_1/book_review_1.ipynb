# Review of Superagency Chapters 1-2

## Introduction

The book starts off with a bit of history on technologies and their role in humanity. They talk about the printing press, the telephone, the Model T, etc and how people have always had skepticism and negative feelings towards these emerging technologies, claiming that they will replace humans in industry or make people less productive in society. In contrast, society has lower unemployment rates and more productivity today. They make claims of how technological progress is inevitable because of the innate collaborative and competitive nature of humans. I thought this was a good opening that set the scene of how people have reacted in the past so that the reader can see the connection with how certain people feel about AI today. 

## Chapter 1: Humanity Has Entered the Chat

Then, in November of 2022 OpenAI released ChatGPT without advanced notice or hype but became popular because of it's unprecedented performance: versatile and incredibly knowledgable. It created a wave of urgency within the tech space because big tech companies felt they fell behind. The government also became skeptical and held hearings on AI.
