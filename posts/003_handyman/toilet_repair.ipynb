# When ChatGPT 5 Fixed a Toilet  
## How My Sister Tightened a Loose Hinge with AI Guidance  

Sometimes AI feels abstract — answering questions about philosophy, coding, or writing. But when my sister showed ChatGPT 5 a picture of her toilet and requested help fixing it, the LLM was able to troubleshoot and lead her to a solution.  

She noticed that the left hinge on her TOTO toilet seat had come loose. Instead of searching through manuals or calling a plumber, she simply snapped a photo and asked ChatGPT: *“The left hinge here is loose, how do I fix?”*  




ChatGPT 5 responded with clear, step-by-step instructions (plus additional information that I omitted): 

1. Lift the hinge cover
2. Locate the screw/bolt
3. Tighten the screw
4. Snap the hinge cover back down

She successfully fixed her toilet!

This small exchange highlights that AI isn’t only for complex reasoning tasks — it can also provide practical, accessible troubleshooting in everyday life.
